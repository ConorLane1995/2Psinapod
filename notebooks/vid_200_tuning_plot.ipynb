```python
import numpy as np
import pickle
import matplotlib as mpl
import matplotlib.pyplot as plt
from cmath import sqrt
```

```python
with open('/home/bic/vtarka/Documents/Vid_202/cells.pkl', 'rb') as f:
        cell_dictionary = pickle.load(f)
```

```python
for cell in cell_dictionary:
    if cell_dictionary[cell]['active']:
        print(cell)
```

```python
CELL_OF_INTEREST = 16
print(cell_dictionary[CELL_OF_INTEREST]['active'])
```

```python
# make a figure with the right number of subplots:
# 1 intensity (1 row)
# 6 frequencies (6 columns)

fig,axs = plt.subplots(6)
this_cells_traces = cell_dictionary[CELL_OF_INTEREST]['traces']

axs_counter = 0
for freq in this_cells_traces:
    for intensity in this_cells_traces[freq]:
        for rep in this_cells_traces[freq][intensity]:
            axs[axs_counter].plot(this_cells_traces[freq][intensity][rep])
            axs[axs_counter].autoscale(enable=True, axis='x', tight=True)
            axs[axs_counter].axvline(x=4,color='k',linestyle='--',linewidth=0.5)
            axs[axs_counter].set_ylim(bottom=0,top=600)
            axs[axs_counter].yaxis.set_visible(False)
            axs[axs_counter].xaxis.set_visible(False)
            axs[axs_counter].text(30,25,freq)
            
    axs_counter +=1

axs[5].yaxis.set_visible(True)
axs[5].xaxis.set_visible(True)
plt.show()
```
