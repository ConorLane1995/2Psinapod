```python
# GET THE MAGNITUDE DIFFERENCE FOR EACH PAIR OF SALINE RECORDINGS.

magnitudes_saline_1  = []

for i in range(len(xoff_all_saline_1[:,0])):

# Assuming x_shifts and y_shifts are lists or arrays containing shift values for each frame
    x_shifts = xoff_all_saline_1[i,:]  # Your x-shift values
    y_shifts = yoff_all_saline_1[i,:]  # Your y-shift values

# Calculate magnitude of motion for each frame (Euclidean distance formula)
    magnitudes = np.sqrt(np.square(x_shifts) + np.square(y_shifts))

    magnitudes_saline_1.append(magnitudes)

magnitudes_saline_1 = np.array(magnitudes_saline_1)

magnitudes_saline_2  = []



for i in range(len(xoff_all_saline_2[:,0])):

# Assuming x_shifts and y_shifts are lists or arrays containing shift values for each frame
    x_shifts = xoff_all_saline_2[i,:]  # Your x-shift values
    y_shifts = yoff_all_saline_2[i,:]  # Your y-shift values

# Calculate magnitude of motion for each frame (Euclidean distance formula)
    magnitudes = np.sqrt(np.square(x_shifts) + np.square(y_shifts))

    magnitudes_saline_2.append(magnitudes)

magnitudes_saline_2 = np.array(magnitudes_saline_2)





```

```python
# GET THE MAGNITUDE DIFFERENCE FOR EACH PAIR OF PSILO RECORDINGS.

magnitudes_psilo_1  = []

for i in range(len(xoff_all_psilo_1[:,0])):

# Assuming x_shifts and y_shifts are lists or arrays containing shift values for each frame
    x_shifts = xoff_all_psilo_1[i,:]  # Your x-shift values
    y_shifts = yoff_all_psilo_1[i,:]  # Your y-shift values

# Calculate magnitude of motion for each frame (Euclidean distance formula)
    magnitudes = np.sqrt(np.square(x_shifts) + np.square(y_shifts))

    magnitudes_psilo_1.append(magnitudes)

magnitudes_psilo_1 = np.array(magnitudes_psilo_1)



magnitudes_psilo_2  = []

for i in range(len(xoff_all_psilo_2[:,0])):

# Assuming x_shifts and y_shifts are lists or arrays containing shift values for each frame
    x_shifts = xoff_all_psilo_2[i,:]  # Your x-shift values
    y_shifts = yoff_all_psilo_2[i,:]  # Your y-shift values

# Calculate magnitude of motion for each frame (Euclidean distance formula)
    magnitudes = np.sqrt(np.square(x_shifts) + np.square(y_shifts))

    magnitudes_psilo_2.append(magnitudes)

magnitudes_psilo_2 = np.array(magnitudes_psilo_2)




```

```python
# Calculate the difference between the MEAN MAGNITUDE SHIFT for pre- and post recordings
mean_magnitude_difference_saline = np.mean(magnitudes_saline_2,axis=1) - np.mean(magnitudes_saline_1,axis=1)
mean_magnitude_difference_psilo = np.mean(magnitudes_psilo_2,axis=1) - np.mean(magnitudes_psilo_1,axis=1)

# Calculate this as a percentage
percent_magnitude_difference_saline = (mean_magnitude_difference_saline/np.mean(magnitudes_saline_1,axis=1))*100
percent_magnitude_difference_psilo = (mean_magnitude_difference_psilo/np.mean(magnitudes_psilo_1,axis=1))*100


#Calculate the difference between the TOTAL DISPLACEMENT for pre- and post-recordings
total_displacement_difference_saline = np.sum(magnitudes_saline_2,axis=1) - np.sum(magnitudes_saline_1,axis=1)
total_displacement_difference_psilo = np.sum(magnitudes_psilo_2,axis=1) - np.sum(magnitudes_psilo_1,axis=1)

# Calculate this as a percentage
percent_displacement_difference_saline = (total_displacement_difference_saline/np.sum(magnitudes_saline_1,axis=1))*100
percent_displacement_difference_psilo = (total_displacement_difference_psilo/np.sum(magnitudes_psilo_1,axis=1))*100
```

```python
# PLOT PERCENTAGE DIFFERENCE IN MEAN SHIFT MAGNITUDE, SALINE DAY AND PSILOCYBIN DAY

mag_saline = pd.DataFrame(percent_magnitude_difference_saline)
mag_psilo = pd.DataFrame(percent_magnitude_difference_psilo)

df_mag = pd.DataFrame(columns=['Saline','Psilocybin'])

df_mag['Saline'] = mag_saline
df_mag['Psilocybin'] = mag_psilo
sns.set_palette("Paired")
ax = sns.boxplot(data=df_mag,showfliers=False)
sns.stripplot(data=df_mag,color='black')

plt.ylabel("% Change")
plt.title("Change in motion shift magnitude",pad=20)


plt.show()
```
