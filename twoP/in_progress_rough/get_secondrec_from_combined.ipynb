```python
import numpy as np
import shutil
import os 
import pickle
import json

JUNCTURE_FRAME = 9000 # frame where recording 1 stops and recording 2 begins
# JUNCTURE_FRAME is included in recording 2, JUNCTURE_FRAME-1 is included in recording 1
```

```python
# load what we need from the config file
#TODO:Make this so that it will find the correct json file on any PC:

with open(os.path.abspath('C:/Users/Conor/2PsinapodConor/config.json')) as f:
    config = json.load(f)

R2_FOLDER = 'C:/Users/Conor/Documents/Imaging_Data/Two-Photon/Psilocybin_Project/Evoked_Cohort_Mice/ID276/ID276_07032023_Psilocybin/Combined/Rec2/' # folder to save the split files for the second recording
BASE_PATH = config['RecordingFolder'] # folder with all of the files generated by Suite2P for this recording (F.npy, iscell.npy, etc)
CSV_PATH = config['Triggers'] # name of CSV (assumed to be in the folder given in line above) with the trigger voltages over the recording
CONDITIONS_PATH = config['Conditions'] # name of the CSV (assumed to be in folder given two lines above) with the condition types of each trial (freq, intensity, etc)
```

```python
def main():
    # load our files that were generated by Suite2P and the stim files
    fluorescence_trace = np.load(BASE_PATH + "F.npy",allow_pickle=True) # uncorrected trace of dF/F
    neuropil_trace = np.load(BASE_PATH + "Fneu.npy",allow_pickle=True) # estimation of background fluorescence
    spks_trace = np.load(BASE_PATH +"spks.npy",allow_pickle=True)


    Fluo2 = fluorescence_trace[:,JUNCTURE_FRAME:]


    Fneu2 = neuropil_trace[:,JUNCTURE_FRAME:]


    Spks2 = spks_trace[:,JUNCTURE_FRAME:]
   
   
    ### SAVING RECORDING 2 INFO ###
    if not os.path.exists(R2_FOLDER):
        os.makedirs(R2_FOLDER)

    np.save(R2_FOLDER + "F.npy",Fluo2)
    np.save(R2_FOLDER + "Fneu.npy",Fneu2)
    np.save(R2_FOLDER + "spks.npy",Spks2)

    # copy some of the shared files into the R2_FOLDER so you don't have to manually move them
    shutil.copy(BASE_PATH+CSV_PATH,R2_FOLDER+CSV_PATH) # copying the trigger file
    shutil.copy(BASE_PATH+CONDITIONS_PATH,R2_FOLDER+CONDITIONS_PATH) # copying the trial labeling

    # copy the Suite2P files that don't need to be split
    shutil.copy(BASE_PATH+"ops.npy",R2_FOLDER+"ops.npy")
    shutil.copy(BASE_PATH+"stat.npy",R2_FOLDER+"stat.npy")
    shutil.copy(BASE_PATH+"iscell.npy",R2_FOLDER+"iscell.npy")

```

```python
if __name__=="__main__":
    main()
```
