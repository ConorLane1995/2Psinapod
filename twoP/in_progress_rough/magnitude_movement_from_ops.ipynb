```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
ops = 'C:/Users/Conor/Documents/Imaging_Data/Two-Photon/Psilocybin_Project/Evoked_Cohort_Mice/ID186/11242022_GCaMP6s_ID186_saline/TSeries-11222022-1228-022/suite2p/plane0/ops.npy'


ops = np.load(ops,allow_pickle=True)
ops = ops.item()
```

```python
print(ops['xoff'])

plt.plot(ops['xoff'])

print(np.mean(ops['xoff']))
```

```python
# Assuming x_shifts and y_shifts are lists or arrays containing shift values for each frame
x_shifts = ops['xoff']  # Your x-shift values
y_shifts = ops['yoff']  # Your y-shift values

# Calculate magnitude of motion for each frame (Euclidean distance formula)
magnitudes = np.sqrt(np.square(x_shifts) + np.square(y_shifts))

# Calculate average magnitude or other statistics
average_magnitude = np.mean(magnitudes)
median_magnitude = np.median(magnitudes)
max_magnitude = np.max(magnitudes)
min_magnitude = np.min(magnitudes)
total_displacement =  np.sum(magnitudes)

print("Average Magnitude:", average_magnitude)
print("Median Magnitude:", median_magnitude)
print("Maximum Magnitude:", max_magnitude)
print("Minimum Magnitude:", min_magnitude)
print("Total displacement:", total_displacement)
```

```python
mag_saline = pd.DataFrame([-13.4425856,-14.8836843,27.31299921,56.07863326,20.33744018,2.377336999,-39.28012678])
mag_psilo = pd.DataFrame([-48.28948504,8.637711939,-7.097653442,43.41231026,-8.267997194,-46.83261411,-33.32102302])

df_mag = pd.DataFrame(columns=['Saline','Psilocybin'])

df_mag['Saline'] = mag_saline
df_mag['Psilocybin'] = mag_psilo
sns.set_palette("Paired")
ax = sns.boxplot(data=df_mag,showfliers=False)
sns.stripplot(data=df_mag,color='black')

plt.ylabel("% Change")
plt.title("Change in motion shift magnitude",pad=20)


plt.show()

# Take the vector of displacement magnitudes and Fourier transform to see if there's a tremor that corresponds to head shake response. 
```
